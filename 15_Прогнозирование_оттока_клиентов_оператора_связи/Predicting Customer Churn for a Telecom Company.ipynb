# Проект: прогнозирование оттока клиентов

Оператору связи необходимо прогнозировать отток клиентов, чтобы предлагать планирующим уйти пользователям промокоды и специальные условия. От оператора предоставлены персональные данные о некоторых клиентах, информацию об их тарифах и договорах.

<b>Цель проекта: </b>на предоставленных данных обучинть несколько моделей и выбрать лучшую для прогнозирования оттока клиентов оператора связи


# Описание услуг

Оператор предоставляет два основных типа услуг:

1. Стационарную телефонную связь. Возможно подключение телефонного аппарата к нескольким линиям одновременно.
2. Интернет. Подключение может быть двух типов: через телефонную линию (DSL*,* от англ. *digital subscriber line*, «цифровая абонентская линия») или оптоволоконный кабель (*Fiber optic*).

Также доступны такие услуги:

- Интернет-безопасность: антивирус (*DeviceProtection*) и блокировка небезопасных сайтов (*OnlineSecurity*);
- Выделенная линия технической поддержки (*TechSupport*);
- Облачное хранилище файлов для резервного копирования данных (*OnlineBackup*);
- Стриминговое телевидение (*StreamingTV*) и каталог фильмов (*StreamingMovies*).

За услуги клиенты могут платить каждый месяц или заключить договор на 1–2 года. Доступны различные способы расчёта и возможность получения электронного чека.


#### Описание данных

Данные состоят из файлов, полученных из разных источников:
 
- информация о договоре;
- персональные данные клиента;
- информация об интернет-услугах;
- информация об услугах телефонии.

Во всех файлах столбец `customerID` содержит код клиента.

Информация о договорах актуальна на 1 февраля 2020.


### Обучение моделей и подбор гиперпараметров

````python
# Функция обучения и предсказания, на вход принимает алгоритм, признаки, целевой признак и параметры
# возвращает словарь с данными о времени обучения и rmse
#def fit_predict(algo, X_train, y_train, X_test, y_test, parametrs):
def fit_predict(algo, X_train, y_train, parametrs):
    
    scoring = {
    'AUC': 'roc_auc'
#    , 
#    'Accuracy': 'accuracy', 
#    'F1': 'f1'
    }

    if algo == 'LinearRegression':
        model_grid = algo()

    else:
        model = algo()

        #model_grid = GridSearchCV(model, parametrs, cv=5, n_jobs=-1, scoring=scoring, refit='AUC')
        model_grid = RandomizedSearchCV(model, parametrs, cv=3, n_jobs=-1, random_state=RANDOM_STATE, scoring=scoring, refit='AUC')
#        model_grid = cross_val_score(model, fit_params=parametrs, cv=5, n_jobs=-1, scoring='neg_root_mean_squared_error')

    start_fit = time.time()
    model_grid_best = model_grid.fit(X_train, y_train)

    stop_fit = time.time()
    start_pred =  time.time()
    predicted = model_grid_best.predict(X_train)
    stop_pred = time.time()
    metrics = model_grid_best.best_score_

    time_fit = stop_fit - start_fit
    time_predict = stop_pred - start_pred
    result_accuracy = accuracy_score(target_train, predicted)
# F-1 мера    
    result_f1 = f1_score(target_train, predicted)
    
    
    t = {'algo' : algo,
         'time_fit': time_fit,
         'time_predict': time_predict,
         'roc_auc': metrics
#         'f1': result_f1,
#         'Accuracy': result_accuracy,
#         'Полнота Recall': recall_score(target_train, predicted),
#         'Точность Precision' : precision_score(target_train, predicted),
#         'Матрица': confusion_matrix(target_train, predicted)
#         'params': model_grid_best.best_params_
        }
    print('Матрица ошибок: \n',confusion_matrix(target_train, predicted))
    print('Доля правильных ответов Accuracy:\n', result_accuracy)
    return t, model_grid_best.best_estimator_.feature_importances_, model_grid_best.best_params_, model_grid_best
````

````python
# Функция рисует ROC-AUC
def ROC_AUC_show(X, y, model):
    fpr, tpr, thresholds = roc_curve(y, model.predict_proba(X)[:, 1])

#    plt.figure(figsize=(8,8))
    plt.plot(fpr, tpr)
    plt.plot([0, 1], [0, 1], linestyle='--')
    plt.xlim([0.0,1.0])
    plt.ylim([0.0,1.0])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC-кривая')
    plt.show()
````

````python
#Функция рисует Признаки по убыванию важности для обучения модели
def impotance_features_plot(model):
# Сделаем словарь с важностью признаков при обучении
    bf = {'algo' : 'CatBoostClassifier',
          'name': features_train.columns,
          'impotances': model.best_estimator_.feature_importances_

         }

    # Отобразим  графически  влияние признака на модель
    (
        pd.DataFrame(bf)
             .sort_values('impotances', ascending=False)
             .plot(kind='bar', y='impotances', x='name')
    )
    plt.xlabel('Название признаков')
    plt.legend(['Признаки'])
    plt.title('Важность признаков')
    plt.show()
````

На предоставленных данных были обучены несколько моделей, и выбрана лучшая по метрике <b>ROC-AUC</b> для прогнозирования оттока клиентов оператора связи

**Описание услуг:**

Оператор предоставляет два основных типа услуг:

* Стационарную телефонную связь. Возможно подключение телефонного аппарата к нескольким линиям одновременно.
* Интернет. Подключение может быть двух типов: через телефонную линию (DSL, от англ. digital subscriber line, «цифровая абонентская линия») или оптоволоконный кабель (Fiber optic).

Также доступны такие услуги:

* Интернет-безопасность: антивирус (DeviceProtection) и блокировка небезопасных сайтов (OnlineSecurity);
* Выделенная линия технической поддержки (TechSupport);
* Облачное хранилище файлов для резервного копирования данных (OnlineBackup);
* Стриминговое телевидение (StreamingTV) и каталог фильмов (StreamingMovies).

За услуги клиенты могут платить каждый месяц или заключить договор на 1–2 года. Доступны различные способы расчёта и возможность получения электронного чека.
