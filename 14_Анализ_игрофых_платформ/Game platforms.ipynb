# Описание проекта

Lоступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). 

Необходимо выявить определяющие успешность игры закономерности, что позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

Предоставлены данные до 2016 года. Для планирования кампании на 2017-й, необходимо отработать принцип работы с данными, чоб можно было прогнозировать продажи на 2017 год по данным 2016-го или же 2027-й — по данным 2026 года.

В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

### Замените названия столбцов (приведите к нижнему регистру)

```python
# Названия столбцов в нижний регистр
data.columns = data.columns.str.lower()
data.columns
```

```python
#Названия игр. платформ, жанров и рейтинга в нижинй регистр
for value in data[['name','platform','genre','rating']]:
    data[value] = data[value].str.lower()
```

###  Преобразуйте данные в нужные типы. Опишите, в каких столбцах заменили тип данных и почему;

```python
# Столбец  Year_of_Release — год выпуска имеет формат float, переделать в целочиленный
data['year_of_release'] = data['year_of_release'].astype('Int64')
data.head()
```

```python
# Посмотрим уникальные значения 
data['critic_score'].sort_values().unique()
```

```python
# Столбец critic_score можно преобразовать в целочисленный тип данных
data['critic_score'] = data['critic_score'].astype('Int64')
data['critic_score'].sort_values().unique()
```

Заодно nan преобразовался в NaN 

```python
data['user_score'].sort_values().unique()
```

В столбце user_score есть значения tbd - аббревиатура от английского To Be Determined (будет определено) или To Be Decided (будет решено). То есть оценки в числовом значении нет, менять на ноль нельзя - нулевой рейтинг есть у игр.
Можно заменить на пропуск (NaN), чтобы преобразовать столбец в float

```python
# Заменим tbd на NaN, тип данных сменит на float
data['user_score'] = data['user_score'].replace('tbd', np.nan).astype('float64')
data['user_score'].sort_values().unique()
```

### •	Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;


Нулевая гипотеза H₀:

- Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.

Альтернативная гипотеза H₁:

- Средние пользовательские рейтинги платформ Xbox One и PC различаются.  

alpha = 0.05 стандартный уровень

```python
data_forecast.query('user_score == "xone"')
```

```python
# выбираем данные по платформам xone и pc
data_forecast_xone = data_forecast.query('platform == "xone" and user_score != -1')['user_score']
data_forecast_pc = data_forecast.query('platform == "pc" and user_score != -1')['user_score']
```

```python
data_forecast_xone.describe()
```

```python
data_forecast_pc.describe()
```

```python
# Делаем t-тест Стьюдента 
# для проверки гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам
# Установка порога значимости
alpha = 0.05
results = st.ttest_ind(
    data_forecast_xone, 
    data_forecast_pc,
    equal_var = False)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Средние пользовательские рейтинги платформ Xbox One и PC одинаковы.
